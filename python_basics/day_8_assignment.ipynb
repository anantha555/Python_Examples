**1. decorators**

```
def takeInput(arg):
    def takerange():
        start = int(input("Enter the starting range number - "))
        end = int(input("Enter the end range number - "))
        arg(start,end)
    return takerange
```

```
@takeInput
def printFibSeries(nterms1, nterms2): 
    n1, n2 = 0, 1
    count = 0
    if nterms1 <= 0:
        print("Please enter a positive integer")
    elif nterms2 == 1:
        print("Fibonacci sequence upto",nterms2,":")
        print(n1)
    else:
        print("Fibonacci sequence:")
        while count < nterms2:
            print(n1)
            nth = n1 + n2
            n1 = n2
            n2 = nth
            count += 1

```

```
printFibSeries()
```

**2. armstrong nm using genarrators**

```
def  armnum(range1 = 100, range2 = 1000):
    for num in range(range1,range2):
        order = len(str(num))
        temp = num
        sum1 = 0
        while temp > 0:
            digit = temp % 10
            sum1 = sum1 + (digit**order)
            temp = temp//10
        if sum1==num:
            #print(sum1)
            yield sum1
```

```
armNumList = armnum(100,1000)
```
