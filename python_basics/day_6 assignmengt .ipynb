**1. Banking using Class**

```
class banking:
  def __init__ (self,ownerName,balance):
    self.ownerName = ownerName
    self.balance = balance
    print("Owner Name      :", self.ownerName,"\nCurrent Balance :",self.balance,"\n")

  def deposite(self, dAmt):
    self.dAmt = dAmt
    self.balance += dAmt
    print("Deposite successful. Updated Balance is :", self.balance)

  def withdraw(self, wAmt):
    self.wAmt = wAmt
    if self.balance >= wAmt:
      self.balance -= wAmt
      print("Withdraw successful. Current Balance is :", self.balance)
    elif self.balance < wAmt:
      print("Insufficiant Balance")  
```

```
p1 = banking("Anantha", 20000000)
p1.deposite(20000000)
p1.withdraw(20000000)
p1.withdraw(40000000)
```

**2. Cone Class**

```
import math
piValue = math.pi

class cone:
  def __init__ (self,R,h):
    self.R = R
    self.h =h
  
  def coneVolume(self):
    v = (1 / 3) * piValue * self.R**2 * self.h
    print("Cone Volume is:", v)

  def coneSurfaceArea(self):
    base = piValue * self.R**2
    side = piValue * self.R * math.sqrt(self.R**2 + self.h**2)
    print("Cone Surface area is:", base+side)
```

```
cone1 = cone(5,12)
print("Height:",cone1.h)
print("Radius:",cone1.R)
cone1.coneVolume()
cone1.coneSurfaceArea()
```
