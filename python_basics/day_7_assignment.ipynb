**2. Prime numberr code with testing**

```
%%writefile prime_numbers.py
'''
this code is to check, a given number is prime or not
'''
def is_prime(num):
    '''
    this code is to check, a given number is prime or not
    '''
    if num <= 1:
        return False
    if num <= 3:
        return True
    if (num % 2 == 0 or num % 3 == 0):
        return False
    i = 5
    while i * i <= num:
        if (num % i == 0 or num % (i + 2) == 0):
            return False
        i = i + 6
    return True

```

```
%%writefile prime_num_testing.py
import unittest
import prime_numbers

class primeTester(unittest.TestCase):
  def testPrimeOne(self):
    temp = 83
    result = prime_numbers.is_prime(temp)
    self.assertAlmostEqual(result,True)
  def testPrimeTwo(self):
    temp = 4
    result = prime_numbers.is_prime(temp)
    self.assertAlmostEqual(result,False)
  def testPrimeThree(self):
    temp = 1
    result = prime_numbers.is_prime(temp)
    self.assertAlmostEqual(result,False)

if __name__ == "__main__":
    unittest.main()
```
